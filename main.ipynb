## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
  print(n, '!')
  if n <= 0:
    return 1

  else:
      local_factorial = n * factorial(n-1)
      return local_factorial
```

```python
#factorial formula
#n! = n * ( n - 1 ) *...*1

# This code defines a function called "factorial" which takes an input "n". The function uses a for loop to iterate through the range of numbers 
# from 1 to n+1. For each number in that range, it multiplies the current value of x by the number in the range. At the end of the loop, 
# the function returns the final value of x, which will be the factorial of the input number "n".

# The Factorial of a positive integer n is the product of all positive integers less than or equal to n. 
# For example, the factorial of 6 (written "6!") is 6 * 5 * 4 * 3 * 2 * 1 = 720.

# So this function takes an input of any positive integer, and returns the factorial of that number.
```

## 7. Write a function that takes a list and returns a list of the unique values.

`NOTE: You cannot use set. 🤔`

```python
def unique(lst_un):
    unique_values = []
    for item in lst_un:
        if item not in unique_values:
            unique_values.append(item)
    return unique_values
```

## 9. Write a function that calculates the standard deviation of a list.
`NOTE: Do not use any libraries or already built functions. 😉`

```python
def st_dev(list_sd):
# Paso 1: Calcular la media
    sum_all = 0
    for num in list_sd:
        sum_all += num
    mean = sum_all/len(list_sd)

 # Paso 2: Calcular la suma de las diferencias al cuadrado
    square_differences = 0
    total_square_differences = 0
    for num in list_sd:
           square_differences = (num - mean)**2
           total_square_differences += square_differences

# Paso 3: Calcular la varianza (promedio de las diferencias al cuadrado)        
    variance = total_square_differences/(len(list_sd)-1)

# Paso 4: Sacar raíz cuadrada de la varianza
    standar_desviation = (variance) **0.5
    
    return standar_desviation
    
 
```

## 10. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once. Mind that the strings may contain characters that are not letters.

```python
def pangram(string):
    text = string.lower()
    letters_found = set()
    for char in text:
        if char.isalpha():
           letters_found.add(char)
    return len(letters_found) == 26
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! 🤪`

```python
def sort_alpha(string):
    # Paso 1: separar manualmente las palabras
    words = []
    current_word = ""
    for char in string:
        if char == ",":
            words.append(current_word)
    #limpiamo current_word para que la funcion pueda seguir leyendo los caracteres        
            current_word = ""
        else:
            current_word += char
    words.append(current_word)  # añadir la última palabra

    # Paso 2: ordenar alfabéticamente
    sorted_words = sorted(words)

    # Paso 3: reconstruir string con comas
    result = ""
    for i in range(len(sorted_words)):
        result += sorted_words[i]
        if i != len(sorted_words) - 1:
            result += ","

    return result
```

```python
# This will test your function 
test_alpha(sort_alpha)
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.

```python
def check_pass(password):
    if len(password) < 8:
        return False  # No cumple con el mínimo de longitud

    has_lower = False
    has_upper = False
    has_digit = False
    has_special = False

    for char in password:
        if char.islower():
            has_lower = True
        elif char.isupper():
            has_upper = True
        elif char.isdigit():
            has_digit = True
        elif not char.isalnum():  # cualquier cosa que NO sea letra ni número
            has_special = True

    # Verificar que todos los requisitos se cumplan
    if has_lower and has_upper and has_digit and has_special:
        return True
    else:
        return False
```

```python
# This will test your function 
test_pass(check_pass)
```
